```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("Customer_data.csv")  # Update with correct file path

# Display dataset information
print("Dataset Info:\n")
df.info()

# Display first few rows
df.head()
```

```python
# Check for missing values
print("Missing Values:\n")
print(df.isnull().sum())

# Check for unique values in categorical columns
print("\nUnique Values in Each Column:\n")
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
# Display rows where TotalCharges is missing
df[df['TotalCharges'].isnull()]
```

```python
# Convert TotalCharges to numeric type
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Fill missing TotalCharges with 0 for customers with tenure = 0
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Verify that missing values are handled
print(df.isnull().sum())  # Should show 0 missing values
```

```python
# Convert Yes/No columns to 0/1
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 
                  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 
                  'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_columns:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
# Ensure SeniorCitizen is an integer
df['SeniorCitizen'] = df['SeniorCitizen'].astype(int)
```

```python
# One-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['InternetService', 'Contract', 'PaymentMethod'], drop_first=True)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the cleaned data
Customer_data = pd.read_csv('Cleaned_Customer_data.csv')

# Drop 'CustomerID' as it's not useful for the model
Customer_data = Customer_data.drop(['customerID'], axis=1)

# Initialize label encoder for binary categorical variables (yes/no, male/female, etc.)
label_encoder = LabelEncoder()

# Encode target variable 'Churn' (binary classification: 0 or 1)
Customer_data['Churn'] = label_encoder.fit_transform(Customer_data['Churn'])

# List of binary categorical columns to encode (yes/no)
binary_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 
                  'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 
                  'Paperless Billing']

# Apply Label Encoding to binary categorical columns
for col in binary_columns:
    if col in Customer_data.columns:
        Customer_data[col] = label_encoder.fit_transform(Customer_data[col])

# For columns with more than two categories, apply One-Hot Encoding
# List of columns to apply One-Hot Encoding (e.g., 'Contract', 'Payment Method', 'Internet Service')
onehot_columns = ['Contract', 'PaymentMethod', 'InternetService']

# Perform One-Hot Encoding on the specified columns
Customer_data = pd.get_dummies(Customer_data, columns=onehot_columns, drop_first=True)

# Check if any non-numeric columns remain
print(Customer_data.dtypes)
```

```python
# Check for any remaining object type columns
print(Customer_data.dtypes)

# Ensure 'Churn' is numeric (if not, convert it)
if Customer_data['Churn'].dtype == 'object':
    Customer_data['Churn'] = label_encoder.fit_transform(Customer_data['Churn'])

# Check if there are any remaining object type columns
print(Customer_data.dtypes)
```

```python
# Check if 'PaperlessBilling' is still an object type
if Customer_data['PaperlessBilling'].dtype == 'object':
    # Apply label encoding for the binary categorical column 'PaperlessBilling'
    Customer_data['PaperlessBilling'] = label_encoder.fit_transform(Customer_data['PaperlessBilling'])

# Check the data types again
print(Customer_data.dtypes)
```
